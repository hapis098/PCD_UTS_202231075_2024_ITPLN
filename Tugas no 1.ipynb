```python
# Import library
import cv2
import matplotlib.pyplot as plt
```

```python
# Membaca citra
img = cv2.imread('citrahafidz.jpg') 
```

```python
# mengkonversi gambar ke warna BGR
img_BGR = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
img_RGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# Memisahkan kanal warna
kanal_merah = img_RGB[:,:,0]
kanal_hijau = img_RGB[:,:,1]
kanal_biru = img_RGB[:,:,2]
```

```python
# Menampilkan gambar kanal warna dalam skala abu-abu
plt.figure(figsize=(15, 4))

# Gambar Asli
plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(img_RGB, cv2.COLOR_RGB2BGR))
plt.title('img Asli')
plt.axis('off')

# Gambar Kanal Merah
plt.subplot(1, 4, 2)
plt.imshow(kanal_merah, cmap='gray')
plt.title('Kanal Merah')
plt.axis('off')

# Gambar Kanal Hijau
plt.subplot(1, 4, 3)
plt.imshow(kanal_hijau, cmap='gray')
plt.title('Kanal Hijau')
plt.axis('off')

# Gambar Kanal Biru
plt.subplot(1, 4, 4)
plt.imshow(kanal_biru, cmap='gray')
plt.title('Kanal Biru')
plt.axis('off')

plt.tight_layout()
plt.savefig('image_kanal_dan_asli.png')
plt.show()
```

```python
# Menghitung histogram untuk setiap kanal warna
hist_merah = cv2.calcHist([kanal_merah], [0], None, [256], [0,256])
hist_hijau = cv2.calcHist([kanal_hijau], [0], None, [256], [0,256])
hist_biru = cv2.calcHist([kanal_biru], [0], None, [256], [0,256])

# Menampilkan histogram
plt.plot(hist_merah, color='r')
plt.plot(hist_hijau, color='g')
plt.plot(hist_biru, color='b')
plt.title('Histogram Warna Merah, Hijau, dan Biru')
plt.xlabel('Nilai Intensitas')
plt.ylabel('Frekuensi')
plt.savefig('histogram_semua_warna.png')
plt.show()
```
